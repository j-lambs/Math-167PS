```python
import numpy as np
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/refs/heads/master/titanic.csv')
```

```python
df.head()
```

# ORDER BY

# SELECTING DATA

```python
df["Name"] # select column by col name
df[0:3] # select rows
```

# Filtering
## 'WHERE' SQL equivalent

```python
df[df["Age"] <= 30]   # WHERE AGE <= 30
```

```python
df[df["Pclass"].isin([1, 2])]   # get only passengers who are pclass = 1 or = 2
```

# Get Percentage of passengers survived

```python
# df["Survived"].mean()
df.groupby("Pclass")["Survived"].mean()
```

# Create a new column FamilySize that is the sum of SibSp and Parch.
## A. SibSp = Number of siblings or spouses aboard the Titanic with the passenger
## B. Parch = Number of parents or children aboard the Titanic with the passenger

```python
df['FamilySize'] = df['SibSp'] + df['Parch']
df
```
